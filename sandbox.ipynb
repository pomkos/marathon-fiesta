```python
import datetime
def avg(dates):
    any_reference_date = datetime.datetime(1900, 1, 1)
    return any_reference_date + sum([date - any_reference_date for date in dates], datetime.timedelta()) / len(dates)
```

```python
df_age = df[(df['Age'] >= 25) & (df['Age'] <=29)]
```

```python
rows = len(df_age)
num = rows * 0.2
miles = 26.2
num
```

```python
from datetime import datetime

def to_minutes(time):
    seconds = (time.hour * 60 + time.minute) * 60 + time.second
    minutes = seconds/60
    return minutes
```

```python
df_age['seconds'] = df_age['Time'].apply(to_seconds)
df_age['minutes'] = df_age['seconds']/60
df_age['miles'] = miles
```

```python
df_age['mpm'] = df_age['minutes']/26.2
```

```python
df_age = df_age.reset_index(drop=True).reset_index()
df_age
```
